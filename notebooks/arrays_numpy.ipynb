# Manipulação de Array com Numpy

```python
import numpy as np
```

```python
# OPERAÇÕES BÁSICAS COM VETORES

# criando vetores
u = np.array([1, 3])
v = np.array([3, 1])
```

```python
# Operações de subtracao e Soma com vetores
y = u - v
z = u + v

print(f"Vetor subtracao: {y}\nVetor soma: {z}")
```

```python
# Fazendo produto escalar e divisao de vetores
a = u / v
b = u * v   # multiplicacao de vetores

print(f"Vetor Divisao: {a}\nVetor Multiplicacao: {b}")

alpha = 2
escalar = alpha * u
print(f"\nProduto escalar de u por alpha: {escalar}")
```

```python
# Fazendo produto interno(escalar) e produto vetorial
interno = np.dot(u, v)

# Vetores tridimensionais
u3 = np.array([1, 3, 1])
v3 = np.array([3, 1, 1])

vetorial = np.cross(u3, v3)

print(f"Resultado produto interno: {interno}\nResultado Produto Vetorial: {vetorial}")
```

# Operações Básicas com Matrizes

```python
# Criando matrizes
I = np.array([[1, 0], [0, 1]])    # Matriz identidade
A = np.array([[2, 1], [1, 2]])
```

```python
B = A + I

print(f"Matriz A:\n {A}\n\nMatriz I:\n {I}\n\nMatriz B = A + I:\n {B}")

B = A - I

print(f"Matriz A:\n {A}\n\nMatriz I:\n {I}\n\nMatriz B = A - I:\n {B}")
```

```python
# Realizando Multiplicação de Matrizes
M = np.matmul(A, I)

print(f"Matriz A:\n {A}\n\nMatriz I:\n {I}\n\nMatriz M = A x I:\n {M}")
```
